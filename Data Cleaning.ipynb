```python
import numpy as np

import pandas as pd
pd.set_option('display.max_columns', 100)

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
```

```python
economic_freedom = pd.read_csv('economic_freedom_index2019.csv', index_col = 'CountryID')
economic_freedom.head()
```

```python
economic_freedom.drop(['WEBNAME','Country'],axis=1, inplace=True)
economic_freedom.head()
```

Removed the WEBNAME and Country columns since they were redundant to the Country Name.

```python
economic_freedom = economic_freedom.rename({'Country Name' : 'Country'}, axis=1)
economic_freedom.head()
```

```python
economic_freedom.isnull().values.any()
```

```python
economic_freedom.isnull().sum().sum()
```

```python
economic_freedom_null = economic_freedom.isnull()
print(economic_freedom_null)
```

```python
economic_freedom = economic_freedom.dropna()
economic_freedom
```

```python
economic_freedom = economic_freedom.drop_duplicates()
economic_freedom
```

```python
economic_freedom.select_dtypes(exclude=['object']).isnull().sum()
```

Checking for as well as counting missing values by feature (numeric)

```python
economic_freedom['2019 Score'].mean()
```

```python
sns.boxplot(economic_freedom['2019 Score'])
plt.show()
```

```python
economic_freedom.loc[economic_freedom['2019 Score'] < 40]
```

When comparing the results of the countries shown with a 2019 score below 40 in this notebook and the one in the other, you see that North Korea and Cuba are not shown here. The reason being is they had NaN values and that was dropped earlier in the notebook.The analysis does better when you have recorded values in all of the column variables to compare and contrast countries rather than apples and oranges.

```python
economic_freedom = economic_freedom[economic_freedom['2019 Score'] > 35]
```

```python
economic_freedom['2019 Score'].mean()
```

```python
sns.boxplot(economic_freedom['2019 Score'])
plt.show()
```

```python
economic_freedom.loc[economic_freedom['2019 Score'] > 85]
```

Even though the countries had no Nan values, these countries were outliers which is why it felt necessary to not count them as well.

```python
economic_freedom = economic_freedom[economic_freedom['2019 Score'] < 85]
```

```python
sns.boxplot(economic_freedom['2019 Score'])
plt.show()
```

```python
economic_freedom['2019 Score'].mean()
```

It seems from removing the outliers on both ends, there's not much of a difference to the mean.
